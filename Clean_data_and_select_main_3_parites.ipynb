## Get Conservatives, Labour and Liberal Democrats candidates infromation
This notebook aims to clean the candidate information data we download, delect irrelevant informations and reduce the parties to decrease the complexity.

```python
import pandas as pd
```

```python
# the data candidate information of 2015 UK GE election has been produced as '2015_candidates.csv'
# df_2015 = pd.read_csv('candidates-2015.csv')
df_2017 = pd.read_csv('candidates-parl.2017-06-08.csv')
```

```python
df_2017.info()
```

```python
df_2017.head(50)
```

```python
df_2017_elected = df_2017[(df_2017['elected'] == True)]
```

```python
len(df_2017_elected)
```

```python
# drop the candidates do not have a twitter account 
df_2017_have_twitter =df_2017[pd.notnull(df_2017['twitter_user_id'])]
```

```python
len(df_2017_have_twitter)
```

```python
df_2017_have_twitter['twitter_user_id'] = df_2017_have_twitter['twitter_user_id'].astype(int)
```

```python
# Drop unimportant columns
df_2017_have_twitter = df_2017_have_twitter.drop(['id', 'honorific_prefix', 'honorific_suffix', 'election', 'party_id', 
                                                 'post_id', 'mapit_url', 'email', 'facebook_page_url', 'party_ppc_page_url',
                                                  'homepage_url', 'homepage_url', 'wikipedia_url', 'linkedin_url', 'image_url',
                                                  'proxy_image_url_template', 'image_copyright', 'image_uploading_user', 
                                                 'election_date', 'election_current', 'party_lists_in_use', 'party_list_position',
                                                  'old_person_ids', 'gss_code' ,'parlparse_id', 'theyworkforyou_url', 'party_ec_id', 
                                                 'favourite_biscuits', 'facebook_personal_url', 'image_uploading_user_notes'], axis=1)
```

```python
df_2017_have_twitter.to_csv('2017_candidates.csv')
```

## SELECTED PARTIES: Conservatives, Labour and Liberal Democrats

```python
df_2017_have_twitter.party_name.unique()
```

```python
Selected_Parties = ['Liberal Democrats', 'Labour Party', 'Conservative and Unionist Party']
```

```python
df_17SP = df_2017_have_twitter[(df_2017_have_twitter['party_name'].isin(Selected_Parties) == True)]
```

```python
len(df_17SP)
```

```python
df_17SP.to_csv('2017_SP_candidates.csv')
```

```python
df_17SP['birth_date']
```

```python
df_17SP.birth_date = df_17SP.birth_date.fillna('2017-00-00')
df_17SP.birth_date = df_17SP.birth_date.apply(lambda x: (2017 - int(x[0:4])))
```
